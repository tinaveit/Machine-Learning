```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
# Question 1

temp = pd.read_csv('ice3.csv')
x = np.array(temp['x'])
y = np.array(temp['y'])
plt.plot(x, y,'o')
```

```python
# Question 2
model = LinearRegression()
x_reshaped = x.reshape((-1,1))
model.fit(x_reshaped,y)
y_pred = model.predict(x_reshaped)

plt.scatter(x_reshaped, y,color='g')
plt.plot(x_reshaped, model.predict(x_reshaped),color='k')

r2 = r2_score(y,y_pred)
r2
```

```python
# Question 3
x_test = np.array([1, 5, 10, 30, 50]).reshape(-1, 1)
y_test_pred = model.predict(x_test)
y_test_pred
```

```python
# Question 4
x_const = sm.add_constant(x)
model_sm = sm.OLS(y, x_const)
results_sm = model_sm.fit()
summary = results_sm.summary()
p_value_slope = results_sm.pvalues[1]

summary, p_value_slope
```

Because our p-value is less than the standard significance of 0.05, we reject the null hypothesis.
